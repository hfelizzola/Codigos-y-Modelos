## Filtrar contratos hospitales públicos

```python
# Cargar datos hospitales
secop_hospitales = pd.read_csv('secop_i_descarga_2020/ips_secop_i.csv.gz',sep=";")
secop_hospitales.head()
```

```python
hospitales_publicos = secop_hospitales.nombre_entidad.unique()
del secop_hospitales
```

```python
# Extraer hospitales

bases_secop = [
    'secop_i_2012.gz',
    'secop_i_2013.gz',
    'secop_i_2014.gz',
    'secop_i_2015.gz',
    'secop_i_2016.gz',
    'secop_i_2017.gz',
    'secop_i_2019.gz',
    'secop_i_2018.gz',
               ]

for i in bases_secop:
  if i == 'secop_i_2012.gz':
    secop_hospitales = pd.read_csv('secop_i_descarga_2020/' + i)
    secop_hospitales = secop_hospitales[secop_hospitales['Nombre de la Entidad'].isin(hospitales_publicos)]
    print("listo " + i)
  else:
    temp = pd.read_csv('secop_i_descarga_2020/' + i)
    temp = temp[temp['Nombre de la Entidad'].isin(hospitales_publicos)]
    secop_hospitales = pd.concat([secop_hospitales, temp])
    print("listo " + i)

#secop_obra['Multa'] = np.int16(secop_obra['UID'].isin(secop_i_match.loc[secop_i_match['Match'] == 'si']['UID'].unique()))
secop_hospitales.to_csv('secop_i_descarga_2020/secop_hospitales.gz', index=False)
```
